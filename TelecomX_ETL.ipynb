#ðŸ”§ TransformaciÃ³n

```python
import numpy as np
```

```python
df.columns=df.columns.str.lower()
df
```

```python
df.columns
```

```python
interest_columns=['churn','customer_gender','customer_seniorcitizen','customer_partner', 'customer_dependents', 'customer_tenure','account_contract','account_paymentmethod',
       'account_charges_monthly', 'account_charges_total']
```

```python
df_transformed=df[interest_columns].copy()
df_transformed
```

```python
df_transformed.info()
```

```python
df_transformed.tail(15)
```

```python
df_transformed.apply(pd.Series.unique)
```

```python
numeric_columns=['customer_tenure','account_charges_monthly', 'account_charges_total']
```

```python
df[numeric_columns].dtypes
```

```python
bool_columns=['churn','customer_seniorcitizen','customer_partner', 'customer_dependents']
```

```python
df['churn'].unique()
```

```python
df_transformed=df_transformed.replace(' ', np.nan)
df_transformed=df_transformed.replace('', np.nan)
```

```python
df_transformed.isna().any()
```

```python
mask_churn=df_transformed['churn'].isna()
df_transformed[mask_churn]
```

```python
rows_nan_churn=df_transformed[mask_churn].index
rows_nan_churn
```

```python
df_clean = df_transformed.dropna(subset=['churn']).copy()
df_clean
```

```python
mask_charges=df_clean['account_charges_total'].isna()
df_clean[mask_charges]
```

```python
rows_nan_charges=df_clean[mask_charges].index
rows_nan_charges
```

```python
months = df_clean['account_contract'].map({'Two year': 24, 'One year': 12})

df_clean['account_charges_total'] = df_clean['account_charges_total'].fillna(
    (df_clean['account_charges_monthly'] * months).round(2)
)
```

```python
df_clean.iloc[rows_nan_charges,9]
```

```python
df_clean[bool_columns]=df_clean[bool_columns].replace({'Yes': 1, 'No': 0})
df_clean
```

```python
df_clean.info()
```

```python
df_clean['account_charges_total']=df_clean['account_charges_total'].astype(np.float64)
```

```python
df_clean[bool_columns]=df_clean[bool_columns].astype(bool)
df_clean.info()
```

```python
daily_accounts = (df_clean['account_charges_monthly'] / 30).round(2)
daily_accounts
```

```python
df_clean.insert(10,'daily_accounts',daily_accounts)
df_clean.head()
```

```python
df_clean.to_json('df_clean_telecomX.json')
```
